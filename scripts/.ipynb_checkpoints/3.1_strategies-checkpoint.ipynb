```python


# ----------------------------------------------------------------------
# EXPECTED INPUT DATAFRAMES
# - summary_filtered: per-channel stats with at least:
#   ['channel_id', 'pol_msgs', 'total_msgs_without_spam', 'pol_ratio_without_spam']
# - df_edges: discovery edges with at least:
#   ['parent', 'type_and_id']  where values look like 'channel_1840578235'
#
# If you already loaded them, keep as-is. Otherwise, uncomment and set paths:
# summary_filtered = pd.read_csv(path_summary_filtered, compression="gzip")
# df_edges = pd.read_csv(path_df_edges)
# ----------------------------------------------------------------------

def strip_channel_prefix(x):
     """'channel_1840578235' -> 1840578235 (int). Return NaN on failure."""
    try:
        return int(str(x).replace('channel_', ''))
    except:
        return np.nan

# --- 1) Normalize keys to integers (safe merges later) -----------------
summary = summary_filtered.copy()
summary['channel_id_num'] = summary['channel_id'].astype(int)

edges = df_edges.copy()
edges['parent_num'] = edges['parent'].apply(strip_channel_prefix)
edges['child_num']  = edges['type_and_id'].apply(strip_channel_prefix)

# --- 2) Attach child metrics to edges (keep edges whose child has stats) ---
child_cols = ['channel_id_num', 'pol_msgs', 'total_msgs_without_spam', 'pol_ratio_without_spam']
child_stats = (summary[child_cols]
               .rename(columns={'channel_id_num': 'child_num'}))

edges_with_child = edges.merge(child_stats, on='child_num', how='inner')

# --- 3) Build the candidate parent list ordered by how political they are ---
parent_stats = (summary[['channel_id_num', 'pol_ratio_without_spam']]
                .rename(columns={'channel_id_num': 'parent_num',
                                 'pol_ratio_without_spam': 'parent_pol_ratio'}))

parents_pool = (edges_with_child[['parent_num']].drop_duplicates()
               .merge(parent_stats, on='parent_num', how='left')
                .dropna(subset=['parent_pol_ratio'])
                .sort_values('parent_pol_ratio', ascending=False)
                .reset_index(drop=True)
               )

total_parents = len(parents_pool)
ordered_parents = parents_pool['parent_num'].toList()

# --- 4) Compute the global total of political messages across ALL children ---
# Use unique children to avoid double counting when a child appears in multiple edges.
unique_children = edges_with_child[['child_num']].drop_duplicates()
pol_msgs_per_child = (child_stats.set_index('child_num')['pol_msgs']
                     .reindex(unique_children['child_num']).fillna(0)
                     )
global_total_pol_msgs = float(pol_msgs_per_child.sum())

if global_total_pol_msgs == 0:
    raise ValueError("Global total of political messages is zero; coverage cannot be computed.")

# --- 5) Choose x-axis steps = fractions of parents explored (10% .. 100%) ---
if total_parents == 0:
    raise ValueError("No candidate parents available after joins/filters.")

fractions = np.linspace(0.1, 1.0, 10)
k_counts = sorted({max(1, int(round(total_parents * f))) for f in fractions})


# --- 6) For each k, compute coverage of political messages (explored vs unexplored) ---
results = []
for k in k_counts:
    chosen_parents = set(ordered_parents[:k])

    # Children under chosen parents = explored children (unique ids)
    explored_mask = edges_with_child['parent_num'].isin(chosen_parents)
    explored_children = edges_with_child.loc[explored_mask, 'child_num'].unique()

    # Children under the remaining parents = unexplored children
    unexplored_children = edges_with_child.loc[~explored_mask, 'child_num'].unique()

    # Compute coverage (share of global political messages)
    explored_share = pol_msgs_per_child.reindex(explored_children).sum() / global_total_pol_msgs * 100
    unexplored_share = 100.0 - explored_share

    results.append({
        'parents_explored_count': k,
        'parents_explored_frac': k / total_parents * 100.0,  # percentage for the x-axis
        'children_explored_count': int(len(explored_children)),
        'children_unexplored_count': int(len(unexplored_children)),
        'explored_coverage_pct': float(explored_share),       # Y (blue)
        'unexplored_coverage_pct': float(unexplored_share),   # Y (red)
    })

coverage_df = pd.DataFrame(results)

# --- 7) Plot coverage curves (Explored vs Unexplored) -------------------
plot_df = coverage_df.melt(
    id_vars=['parents_explored_frac'],
    value_vars=['explored_coverage_pct', 'unexplored_coverage_pct'],
    var_name='Group',
    value_name='Coverage (%)'
)

# Human-friendly legend labels
plot_df['Group'] = plot_df['Group'].map({
    'explored_coverage_pct': 'Explored children',
    'unexplored_coverage_pct': 'Unexplored children'
})

fig = px.line(
    plot_df,
    x='parents_explored_frac',
    y='Coverage (%)',
    color='Group',
    markers=True,
    labels={'parents_explored_frac': 'Fraction of parents explored (%)'}
)
fig.update_layout(
    title='Coverage of Political Messages vs. Fraction of Parents Explored',
    xaxis=dict(tickformat='.0f'),
    yaxis=dict(range=[0, 100])
)
fig.show()

# Optional: quick table preview
display_cols = [
    'parents_explored_frac',
    'parents_explored_count',
    'children_explored_count',
    'children_unexplored_count',
    'explored_coverage_pct',
    'unexplored_coverage_pct'
]
print(coverage_df[display_cols].round(2).to_string(index=False))


```
